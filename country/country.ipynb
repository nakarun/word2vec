## ベクトル間のcos類似度を求めてみる

参考 : http://emoson.hateblo.jp/entry/2015/02/21/234311

以下は失敗

model['東京']　+ model['フランス'] - model['日本'] 

```python
type(model['東京'])
```

```python
a = model['東京']
```

```python
b = model['フランス']
```

```python
c = model['日本']
```

```python
d = model['パリ']
```

```python
left_1 = a + b - c
```

```python
left_2 = a + b - d
```

```python
def cosine_similarity(v1, v2):

    return sum([a*b for a, b in zip(v1, v2)])/(sum(map(lambda x: x*x, v1))**0.5 * sum(map(lambda x: x*x, v2))**0.5)
```

### 東京 + フランス - 日本 = パリ

```python
cosine_similarity(left_1, d)
```

### 東京 + フランス - パリ = 日本

```python
cosine_similarity(left_2, c)
```
